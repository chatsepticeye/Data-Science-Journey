```python
# Move everything back to columns temporarily for the calculation
data.reset_index().groupby('gender')['age'].max()
```

```python
# 1. Move everything back to regular columns
df_clean = data.reset_index()

# 2. Force the 'name' and 'age' columns to be clean (no spaces)
df_clean['name'] = df_clean['name'].astype(str).str.strip()
df_clean['age'] = pd.to_numeric(df_clean['age'], errors='coerce')

# 3. Now search normally without the index headache
result = df_clean[df_clean['name'] == 'jia']
print(result)
```

```python
# 1. Reset the index so we aren't fighting the MultiIndex rules
df_temp = data.reset_index()

# 2. Use .astype(str) so we can find Jia even if there are hidden spaces
result = df_temp[df_temp['name'].astype(str).str.contains('jia')]

print(result)
```
