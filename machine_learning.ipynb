```python
import pandas as pd
```

```python
data=pd.read_csv(r'C:\Users\Chaitanya\Downloads\50startupsdataset.zip')
```

```python
data
```

```python
data=data.iloc[:,1:]
```

```python
data
```

```python
data.drop('State',axis=1,inplace=True)
```

```python
#dividing data in to output and input
x=data.iloc[:,0:3]
y=data.iloc[:,3]
```

```python
#dividing data into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.80)
```

```python
data.shape
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
#creating a blank model 
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#training the model with xtrain and ytrain
model.fit(xtrain,ytrain)
```

```python
#making the prediction from xtest
yp=model.predict(xtest)
```

```python
#comparing the actual outout and predicted output
from sklearn.metrics import r2_score
r2_score(ytest,yp)
```
