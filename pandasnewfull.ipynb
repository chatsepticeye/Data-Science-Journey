```python
##concatenate and merge dataframe(JOINS)
df1=pd.DataFrame({'ID':[1,2,3],'Name':['A','B','C']})
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,2,4],'Score':['90','77','88','99']})
df2
```

```python
pd.concat([df1,df2],axis=0) #vertical / top on top / row level join
```

```python
pd.concat([df1,df2],axis=1) #horizontal / side by side / column level join
```

```python
pd.merge(df1,df2,how='inner',on='ID')
```
